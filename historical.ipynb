```python
cagr = 0.0686
nyears = 2013 - 1870
```

```python
x = np.linspace(1871,2013,nyears)
n = np.linspace(1,nyears,nyears)
y = (1+cagr)**n
```

```python
real_growth = []
for row in dict_reader:
    real_growth.append(row['growth'].strip())
```

```python
real_g = np.array(real_growth)
real_g = real_g[::-1]
real_g[3]
```

```python
product_g = 1
for i in range(1, len(real_g)):
    product_g *= (1 + float(real_g[i-1])/100)
geo_mean = (product_g**(1.0/len(real_g)) - 1) *100.0
geo_mean
```

### oh, it's not the same as the number given by the site, which probably means that the site is actually doing a fit?
#### i will recreate the fit

```python
len(real_g)
real_y = [1]
for i in range(1, len(real_g)):
    real_y.append(real_y[i-1]*(1+float(real_g[i-1])/100))
real_y = np.array(real_y)
```

```python
# log y = a + bx
real_log = np.log(real_y)
slope, intercept, r_value, p_value, std_err = stats.linregress(n,real_log)
# standard error is 0.0006894
np.exp(std_err)
slope
```

```python
# now assume no intercept
# logy = Ap where A = [x] and p = [b]
A = np.vstack([n, np.zeros(len(n))]).T
slope, intercept = np.linalg.lstsq(A,real_log)[0]
slope
# 0.0664 
# no intercept is the winner here
```

```python
# i wanna get the sample standard dev
real_var = np.sum((real_log - n*slope)**2)/len(n)
real_stdev = real_var**(0.5)
diffs = real_log - n*slope
diffs2 = diffs**2
ppl.plot(n,diffs, linewidth = 3.0)
ppl.plot(n,diffs2, linewidth = 3.0)
real_stdev
```

# take a segment
#### make assumption that noise represents mispricing and thus misallocation of capital
#### then more noise should cause average growth to be lower

```python
# s for segment
xs = x[100:130]
real_ys = real_y[100:130]
ys = y[100:130]
```

```python
ppl.plot(xs, real_ys, linewidth = 3.0)
ppl.plot(xs,ys,linewidth = 3.0)
```

```python
# logy = intercept + slope*x + std_err
noisy_logs1 = []
noisy_ys1 = []
for i in range(100,130):
    l = np.random.normal(slope*i, real_stdev, 1)[0]
    noisy_logs1.append(l)
    noisy_ys1.append(np.exp(l))    
noisy_logs1 = np.array(noisy_logs1)
noisy_ys1 = np.array(noisy_ys1)
```

```python
# y with more noise
noisy_logs2 = []
noisy_ys2 = []
for i in range(100,130):
    l = np.random.normal(slope*i, real_stdev*1.5, 1)[0]
    noisy_logs2.append(l)
    noisy_ys2.append(np.exp(l))    
noisy_logs2 = np.array(noisy_logs2)
noisy_ys2 = np.array(noisy_ys2)
```

```python
ppl.plot(xs, real_ys, linewidth = 3.0)
ppl.plot(xs,ys,linewidth = 3.0)
ppl.plot(xs,noisy_ys1,linewidth = 3.0)
ppl.plot(xs,noisy_ys2,linewidth = 3.0)
```

```python
# compare error of noisy_ys2 and real_ys
real_e = np.sum((real_ys - ys)**2)**(0.5)
real_e
e2 = np.sum((noisy_ys2 - ys)**2)**(0.5)
(e2 - real_e)/e2
```

```python
# does more noise necessarily leads to less slope anyway? no it shouldn't
# compute geometric mean of noisy_ys2
noisy_growth2 = []
for i in range(1,30):
    noisy_growth2.append(noisy_ys2[i]/noisy_ys2[i-1])
noisy_growth2 = np.array(noisy_growth2)
np.prod(noisy_growth2)**(1.0/30)
```

```python
# p% more standard dev leads to some q% less log_slope
# it's a guess at a relationship assuming that misallocation of capital degrades market growth
# take a guess that q is p/10
# for 50% more standard dev then, log_slope is 5% less
# y with more noise

# changing the slope changes the intercept
# calculate new intercept so that y[100] = (1+slope)^100*e^intercept
low_slope = slope*0.95
logi3 = np.log(y[100])-np.log((1+low_slope)**100)
logi3
```

```python
noisy_logs3 = []
noisy_ys3 = []
ys3 = []
for i in range(100,130):
    l = np.random.normal(low_slope*i, real_stdev*1.5, 1)[0]+logi3
    noisy_logs3.append(l)
    noisy_ys3.append(np.exp(l))
    ys3.append(y[100]*(1+low_slope)**(i-100))
noisy_logs3 = np.array(noisy_logs3)
noisy_ys3 = np.array(noisy_ys3)

noisy_growth3 = []
for i in range(1,30):
    noisy_growth3.append(noisy_ys3[i]/noisy_ys3[i-1])
noisy_growth3 = np.array(noisy_growth3)
np.prod(noisy_growth3)**(1.0/30.0)
```

```python
ppl.plot(xs, real_ys, linewidth = 3.0)
ppl.plot(xs,ys,linewidth = 3.0)
ppl.plot(xs,noisy_ys3,linewidth = 3.0)
ppl.plot(xs,ys3,linewidth = 3.0)
```
