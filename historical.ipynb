```python
import csv
```

```python
dict_reader = csv.DictReader(open("annual growth.csv"))
```

```python
real_growth = []
for row in dict_reader:
    real_growth.append(row['growth'].strip())
```

```python
real_g = np.array(real_growth)
real_g = real_g[::-1]
real_g[3]
```

```python
product_g = 1
for i in range(1, len(real_g)):
    product_g *= (1 + float(real_g[i-1])/100)
geo_mean = (product_g**(1.0/len(real_g)) - 1) *100.0
geo_mean
```

### oh, it's not the same as the number given by the site, which probably means that the site is actually doing a fit.
#### i should try to recreate the fit

```python
len(real_g)
real_y = [1]
for i in range(1, len(real_g)):
    real_y.append(real_y[i-1]*(1+float(real_g[i-1])/100))
real_y = np.array(real_y)
```

```python
ppl.plot(x,real_y, linewidth = 3.0)
ppl.plot(x,y,linewidth = 3.0)
```

# take a segment
#### make assumption that noise represents mispricing and thus misallocation of capital
#### then more noise should cause average growth to be lower

```python
# s for segment
xs = x[100:130]
real_ys = real_y[100:130]
ys = y[100:130]
```

```python
ppl.plot(xs, real_ys, linewidth = 3.0)
ppl.plot(xs,ys,linewidth = 3.0)
```

```python
# y with more noise
# standard dev of annual growth is 18.68%
# just estimate noise based on baseline gs 
# standard devs of growth will be constant
noisy_gs1 = []
noisy_ys1 = [ys[0]]
for i in range(1,30):
    g = np.random.normal(0.0855, 0.18, 1)[0]
    noisy_gs1.append(g)
    noisy_ys1.append(noisy_ys1[i-1]*(1+g))    
noisy_gs1.append(0.068)
noisy_gs1 = np.array(noisy_gs1)
noisy_ys1 = np.array(noisy_ys1)
```

```python
# another way would be to multiply error real_gs - 0.0686 by some number
real_gs = real_g[100:130]
noisy_gs2 = []
noisy_ys2 = [real_ys[0]]
for i in range(1,30):
    g = 0.0686 + (float(real_gs[i-1])/100.0 - 0.0686)*1.6
    noisy_gs2.append(g)
    noisy_ys2.append(noisy_ys2[i-1]*(1+g))
noisy_gs2.append(0.0685)
noisy_gs2 = np.array(noisy_gs2)
noisy_ys2 = np.array(noisy_ys2)
# this is bad
# need log
```

```python
ppl.plot(xs, real_ys, linewidth = 3.0)
ppl.plot(xs,ys,linewidth = 3.0)
ppl.plot(xs,noisy_ys1,linewidth = 3.0)
ppl.plot(xs,noisy_ys2,linewidth = 3.0)
```

```python
ppl.plot(xs, real_gs, linewidth = 3.0)
ppl.plot(xs,100.0*noisy_gs1,linewidth = 3.0)
ppl.plot(xs,100.0*noisy_gs2,linewidth = 3.0)
```
